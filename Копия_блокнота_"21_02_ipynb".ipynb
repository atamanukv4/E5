```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4*np.pi,100)
plt.figure()
plt.plot(x, np.sin(x), 'r-.')
plt.plot(x,np.cos(x),'b:')
plt.title('sin x  cos x')
plt.grid()
plt.show()
```
